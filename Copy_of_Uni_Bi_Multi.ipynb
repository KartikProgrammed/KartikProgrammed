# Univariate Analysis

```python
#Univariate Analysis is a type of data visualization where we visualize only a single variable at a time.
#Univariate Analysis helps us to analyze the distribution of the variable present in the data so that we can perform further analysis
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
data = pd.read_csv('Employee_dataset.csv')
sns.histplot(data['age'])
```

```python
sns.countplot(data['gender_full'])
```

```python
data.head()
```

```python
data.describe()
```

```python
x = data['STATUS_YEAR'].value_counts()
plt.pie(x.values, labels=x.index, autopct='%1.1f%%')
plt.show()
```

# Bivariate Analysis

```python
#Bivariate analysis is the simultaneous analysis of two variables.
#It explores the concept of the relationship between two variable whether there exists an association
#and the strength of this association or whether there are differences between two variables and the significance of these differences.

#Categorical v/s Numerical
plt.figure(figsize=(15, 5))
sns.barplot(x=data['department_name'], y=data['length_of_service'])
plt.xticks(rotation='vertical')
```

```python
#Numerical v/s Numerical
sns.scatterplot(x=data['length_of_service'], y=data['age'])
```

```python
#Categorical v/s Categorical
sns.countplot(x=data['STATUS_YEAR'], hue=data['STATUS'])
```

# Multivariate Analysis

```python
#It is an extension of bivariate analysis which means it involves multiple variables at the same time to find correlation between them.
#Multivariate Analysis is a set of statistical model that examine patterns in multidimensional data by considering at once,
#several data variable.
from sklearn import datasets, decomposition
iris = datasets.load_iris()
X = iris.data
y = iris.target
pca = decomposition.PCA(n_components=2)
X = pca.fit_transform(X)
sns.scatterplot(x=X[:, 0], y=X[:, 1], hue=y)
```
